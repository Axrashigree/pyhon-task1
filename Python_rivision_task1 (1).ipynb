# Task 1: NumPy Basics

Write a Python script that creates a NumPy array with random integers and performs basic mathematical operations like addition, subtraction, multiplication, and division on it.

```python
import numpy as np

# Generate random array of size 3 with integers between 1 and 10
arr = np.random.randint(1, 11, 3)
print(" Array:", arr)

# Addition
print("Addition:", arr + 5)

# Subtraction
print("Subtraction:", arr - 3)

# Multiplication
print("Multiplication:", arr * 2)

# Division
# Here, we'll use np.where to avoid dividing by zero
print("Division:", np.where(arr != 0, arr / 2, 0))
```

Task 2: Pandas Data Analysis

Using Pandas, load a dataset of your choice (you can find datasets online or use your own). Clean the data and perform basic data analysis tasks like calculating mean, median, and mode for specific columns.

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("D:/DSAI/HKKH.csv")

# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())


# Assuming there are missing values (there aren't in this dataset),
# you could fill them using the mean of the respective column as follows:
df.fillna(df.mean(), inplace=True)

# Calculate mean, median, and mode for the 'sepal_length' column
mean_val = df['mag'].mean()
median_val = df['mag'].median()
mode_val = df['mag'].mode().iloc[0]  # mode() returns a Series, we extract the first value

print("\nAnalysis for 'mag' column:")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")
```

Task 3: Data Visualization
Using Matplotlib, create a bar chart or histogram to visualize the distribution of a dataset of your choice. Label the axes and add a title to the chart.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("D:/DSAI/HKKH.csv")

# Calculate the mean mag length for each place
place = df.groupby('place')['mag'].mean()

# Create a bar chart to visualize the mean mag length for each place
plt.figure(figsize=(18, 6))
place.plot(kind='bar', color=['Blue', 'black', 'purple'])
plt.xlabel('place')
plt.ylabel('mag')
plt.title('Place vise mag')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Task 4: NumPy Array Manipulation
Create two NumPy arrays and perform operations to concatenate them vertically and horizontally.


```python
import numpy as np

# Create two NumPy arrays
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[7, 8, 9], [10, 11, 12]])

print("Array 1:")
print(array1)

print("\nArray 2:")
print(array2)

# Concatenate arrays vertically
vertical_concat = np.vstack((array1, array2))
print("\nVertical Concatenation:")
print(vertical_concat)

# Concatenate arrays horizontally
horizontal_concat = np.hstack((array1, array2))
print("\nHorizontal Concatenation:")
print(horizontal_concat)
```

Task 7: NumPy Statistical Analysis
Use NumPy to calculate descriptive statistics like variance, standard deviation, and correlation coefficients for a dataset of your choice (you can find datasets online or use your own).


```python
import numpy as np
import seaborn as sns

# Load the dataset 
df = pd.read_csv("D:/DSAI/HKKH.csv")

# Extract the depth and depthError columns
depth = df['depth'].values
depthError = df['depthError'].values

# Calculate variance
variance_depth = np.var(depth)
variance_depthError = np.var(depthError)

# Calculate standard deviation
std_dev_depth = np.std(depth)
std_dev_depthError = np.std(depthError)

# Calculate correlation coefficient between depth and depthError
corr_coefficient = np.corrcoef(depth, depthError)[0, 1]

# Display results
print(f"Variance of depth: {variance_depth:.2f}")
print(f"Variance of depthError: {variance_depthError:.2f}")
print(f"\nStandard Deviation of depth: {std_dev_depth:.2f}")
print(f"Standard Deviation of depthError: {std_dev_depthError:.2f}")
print(f"\nCorrelation Coefficient between depth and depthError: {corr_coefficient:.2f}")
```

Task 10: Pandas Data Visualization Choose a dataset of your choice and create various visualizations (e.g., bar charts, pie charts, scatter plots) using Pandas to represent different aspects of the data, such as trends or distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("D:/DSAI/HKKH.csv")

# Bar Chart: Display the mean of mag for each place
data.groupby('place')['mag'].mean().plot(kind='bar', color=['red', 'green', 'blue'])
plt.title('Average mag by place')
plt.ylabel('Average mag')
plt.show()

# Pie Chart: Distribution of place in the dataset
data['place'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of place')
plt.show()

# Scatter Plot: mad vs.depth
data.plot(kind='scatter', x='mag', y='depth', c='depthError', colormap='viridis', s=50)
plt.title('Scatter Plot: mag vs. depth(Color coded by depthError)')
plt.show()

# Histogram: Distribution of depthError
data['depthError'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of depthError')
plt.xlabel('depthError')
plt.show()


```
